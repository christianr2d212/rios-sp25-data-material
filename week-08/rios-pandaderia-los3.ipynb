applying a function to change the shortened letters in marrystatus to full letters

```python
def replacus_msw(some_string):
    if some_string== 'm':
        return "married"
    elif some_string=='s':
        return "single"
    else:
        return 'widowed'


        
```

```python
utah_df['marrystatus'] = utah_df['marrystatus'].apply(replacus_msw)
```
